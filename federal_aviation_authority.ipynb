```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/faa_ai_prelim.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
data=df[['ACFT_MAKE_NAME','LOC_STATE_NAME','ACFT_MODEL_NAME','RMK_TEXT','FLT_PHASE','EVENT_TYPE_DESC','FATAL_FLAG']]
```

```python
data.head()
```

```python
data['FATAL_FLAG'].fillna(value='No', inplace=True)
```

```python
data=data.dropna(subset=['ACFT_MAKE_NAME'])#drop values where aircraft make name is not available
```

```python
data.shape
```

```python
acft_type=data.groupby('ACFT_MAKE_NAME')#group by aircraft name
acft_type.size()
```

```python
fatal_acc=data.groupby('FATAL_FLAG')#group by fatal accidents
fatal_acc.size()
```

```python
acc_with_fatality=fatal_acc.get_group('Yes')#selecting accidents with fatality flag yes
acc_with_fatality
```
